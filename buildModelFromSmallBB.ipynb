```python
# Load CSV data into df
df_data = pd.read_csv ('/content/drive/MyDrive/labels.csv')
df_data
```

```python
# only use data with high confidence
df_filtered = df_data[df_data.conf > 0.6]
df_filtered = df_filtered.drop(['filename', 'width', 'height', 'id_big', 'xmin', 'ymin', 'xmax', 'ymax','xmin2', 'ymin2', 'xmax2', 'ymax2'], axis=1)
df_filtered
```

```python

for index,row in df_filtered.iterrows():
  (conf, unicod) = row
  unicod = str(unicod).replace("/", "~")
  smallCropFile = os.path.join("/content/drive/MyDrive/OCR/crop_small",(str(unicod)+"_"+str(conf)+".png"))
  print(smallCropFile)
  image = cv2.imread(smallCropFile)
  lenU = len(unicod)
  (h,w,c) = image.shape
  if(h>w):
    image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
    (h,w,c) = image.shape
  im_number = 0
  width_cut = w // lenU
  for i in range(lenU):
    cuttedIm = image[0:h,(im_number*width_cut):((im_number+1)*width_cut)]
    cv2.imwrite('im_{}.png'.format(im_number), cuttedIm)
    im_number += 1
```
